```python
def barneys_scraper1(num_pages, link, Category):
    #pages to loop over
    pages = [str(i) for i in range(1, num_pages+1)]
    #initializing lists to put scraped data into
    brands=[]
    items=[]
    prices=[]
    #looping over pages
    for page in pages:
        url = link+'?page='+page
        #incorporating Selenium to interact with webpage
        driver = webdriver.Chrome()
        driver.implicitly_wait(30)
        driver.get(url)
        #accounts for popup not always appearing
        try:
            #telling Selenium where it can find the button and actually clicking it
            button = driver.find_element_by_css_selector("button[type='reset']")
            button.click()
        except:
            pass
        #bringing BS into the equation to actually scrape
        soup_level1 = BeautifulSoup(driver.page_source, 'html.parser')
        #getting down to only the code that I'm interested in
        body = soup_level1.body.find('div', id='atg_store_catSubProdList').find(
            'div', id='atg_store_prodList').ul.find_all('li')
        #looping over each iteration, scraping and appending to lists created earlier
        for b in body:
            brand = b.find('div', class_='brand').text.strip()
            name = b.find('div', class_='product-name').text.strip()
            price = b.find('span', class_='product-standard-price').text.strip()
            brands.append(brand)
            items.append(name)
            prices.append(price)
        #quitting out of Selenium
        driver.quit()
    #storing the results in a Pandas dataframe
    return pd.DataFrame({'Brand':brands, 'Product_Name':items, 'Price':prices, 'Category':Category})
```

```python
def barneys_scraper2(num_pages, link, Category):
    pages = [str(i) for i in range(1, num_pages+1)]
    brands=[]
    items=[]
    prices=[]
    for page in pages:
        url = link+'?page='+page
        driver = webdriver.Chrome()
        driver.implicitly_wait(30)
        driver.get(url)
        try:
            button = driver.find_element_by_css_selector("button[type='reset']")
            button.click()
        except:
            pass
        soup_level1 = BeautifulSoup(driver.page_source, 'html.parser')
        body = soup_level1.body.find('div', id='atg_store_catSubProdList').find(
            'div', id='atg_store_prodList').div.find_all('li')
        for b in body:
            brand = b.find('div', class_='brand').text.strip()
            name = b.find('div', class_='product-name').text.strip()
            price = b.find('span', class_='product-standard-price').text.strip()
            brands.append(brand)
            items.append(name)
            prices.append(price)
        driver.quit()
    return pd.DataFrame({'Brand':brands, 'Product_Name':items, 'Price':prices, 'Category':Category})
```
