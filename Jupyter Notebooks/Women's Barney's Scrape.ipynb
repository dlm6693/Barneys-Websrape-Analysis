```python
from bs4 import BeautifulSoup
from selenium import webdriver
import re
import pandas as pd
import os
import requests
```

```python
def barneys_scraper(num_pages, link, Category):
    pages = [str(i) for i in range(1, num_pages+1)]
    brands=[]
    items=[]
    prices=[]
    for page in pages:
        url = link+'?page='+page
        driver = webdriver.Chrome()
        driver.implicitly_wait(30)
        driver.get(url)
        try:
            button = driver.find_element_by_css_selector("button[type='reset']")
            button.click()
        except:
            pass
        soup_level1 = BeautifulSoup(driver.page_source, 'html.parser')
        body = soup_level1.body.find('div', id='atg_store_catSubProdList').find(
            'div', id='atg_store_prodList').div.find_all('li')
        for b in body:
            brand = b.find('div', class_='brand').text.strip()
            name = b.find('div', class_='product-name').text.strip()
            price = b.find('span', class_='product-standard-price').text.strip()
            brands.append(brand)
            items.append(name)
            prices.append(price)
        driver.quit()
    return pd.DataFrame({'Brand':brands, 'Product_Name':items, 'Price':prices, 'Category':Category})
```

```python
women_coats = barneys_scraper(5, 'https://www.barneys.com/category/women/clothing/coats/N-ih1kp', 'Coats')
women_coats.to_csv('womencoats.csv')
women_coats.head()
```
